```python
import sys, platform

print("Python:", sys.version)
print("Platform:", platform.platform())
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 200)
y = np.sin(x)

plt.plot(x, y)
plt.title("sine wave")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```
